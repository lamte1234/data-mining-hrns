### 8, Normalize the data AGE=[25, 56, 63, 32, 41, 49] ; Income=[49000, 156000, 99000, 192000, 39000, 57000]

```python
age = [25, 56, 63, 32, 41, 49]
income = [49000, 156000, 99000, 192000, 39000, 57000]
def normalize_min_max(a):
    temp = []
    for elem in a:
        temp += [(elem - min(a)) / (max(a) - min(a)) ]
    return temp
print(normalize_min_max(age))
print(normalize_min_max(income))
```
